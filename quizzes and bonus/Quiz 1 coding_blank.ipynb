### Probelm description:

You are given a dataset containing information about the number of hours studied by students and their corresponding exam scores. Your task is to build a simple linear regression model to predict exam scores based on the number of hours studied.


```python
# Don't modify this chunck of code
import pandas as pd

# The given Data set df
data = {'Hours_Studied': [2, 3, 5, 7, 8, 10, 11, 12],
        'Exam_Score': [65, 70, 80, 85, 90, 95, 97, 100]}
df = pd.DataFrame(data)
```

### Questions
1. **Import Libaries:** Import the necessary libraries for this task. You'll need pandas for data manipulation, matplotlib.pyplot for plotting, and sklearn.linear_model for linear regression.

```python
# Please Import the necessary libraries
# Your code here
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
```

2. **Data Exploration:**

* Print the first 5 rows of the DataFrame df
* Create a scatter plot of 'Hours_Studied' (x-axis) vs. 'Exam_Score' (y-axis). Briefly describe the relationship you observe.

```python
df.head()
```

```python
# Your code here:
plt.scatter(data['Hours_Studied'], data['Exam_Score'])
plt.xlabel('Hours_Studied')
plt.ylabel('Exam_Score')
plt.title('Scatter Plot of Hours Studied versus Exam Score')
plt.show()
df.head()
```

### Description

- As the number of hours studied increases, the exam score tends to increase. Linear regression seems fine.

3. **Prepare the Data:**
* Create two variables:
    * X: A DataFrame containing the 'Hours_Studied' column (independent variable).
    * y: A Series containing the 'Exam_Score' column (dependent variable).

```python
# Your code here:
# Prepare the Data

X = pd.DataFrame(data)['Hours_Studied'].values.reshape(-1, 1)

y = pd.DataFrame(data)['Exam_Score'].values
```

4. **Create and Train the Model:**

* Initialize a Linear Regression model from sklearn.linear_model.
* Train the model using the X and y data you prepared.

```python
# Your code here:
model = lm.LinearRegression()
model.fit(X, y)
```

5. **Confidence Interveal of the predictors**
* Please report $\beta_0$ and $\beta_1$ and write down the estimated linear model

```python
# Your code here:
print("Beta 0 =", model.intercept_)
print("Beta 1 =", model.coef_[0])
```

6. **Make Predictions:**

* Use the trained model to predict the exam score for a student who studied for 6 hours.

```python
# Your code here:
hours = 6
predicted_score = model.predict([[hours]])

print(f"Exam Score after {hours} hours studied is: {predicted_score[0]:.3f}")
```
